```python
%%writefile team4_app.py
# All Imports
###
import streamlit as st
import pandas as pd
import numpy as np
import os
import sys

# Title for Webpage
###
st.title('Welcome to MonoHuman!')
###

# Subsection for what MonoHuman is
###
st.header('About MonoHuman:')
st.write("MonoHuman is an Animatable Human Neural Field from Monocular Video")

st.image('mohu.jpeg', caption='MonoHuman', width=None)

# Background on MonoHuman
with st.expander("Learn about MonoHuman"):
    st.write("Animating virtual avatars with free-view control is crucial for various applications like virtual reality and digital entertainment. Previous studies have attempted to utilize the representation power of the neural radiance field (NeRF) to reconstruct the human body from monocular videos. Recent works propose to graft a deformation network into the NeRF to further model the dynamics of the human neural field for animating vivid human motions. However, such pipelines either rely on pose-dependent representations or fall short of motion coherency due to frame-independent optimization, making it difficult to generalize to unseen pose sequences realistically. MonoHuman, which robustly renders view-consistent and high-fidelity avatars under arbitrary novel poses. Our key insight is to model the deformation field with bi-directional constraints and explicitly leverage the off-the-peg keyframe information to reason the feature correlations for coherent results. Specifically, we first propose a Shared Bidirectional Deformation module, which creates a pose-independent generalizable deformation field by disentangling backward and forward deformation correspondences into shared skeletal motion weight and separate non-rigid motions. Then, we devise a Forward Correspondence Search module, which queries the correspondence feature of keyframes to guide the rendering network. The rendered results are thus multi-view consistent with high fidelity, even under challenging novel pose settings. Extensive experiments demonstrate the superiority of our proposed MonoHuman over state-of-the-art methods.")


# Create subsection for Demo Video
###
option = st.selectbox('What would you like to see',
                      ('Freeview', 'Movement', 'Action', 'Text to Action'), index=None,
   placeholder="Select the method...")

if option == 'Freeview':
    st.header('Demo Video for FreeView:')
    st.video('freeview.mp4')
    st.caption('This is a Demo Video displaying the rendering result and motion quality we have generated.')

elif option == 'Movement':
    st.header('Demo Video for Movement:')
    st.video('movement.mp4')
    st.caption('This is a Demo Video displaying the rendering result and motion quality we have generated.')

elif option == 'Action':
    st.header('Demo Video for Action:')
    st.video('text_animation.mp4')
    st.caption('This is a Demo Video displaying the rendering result and motion quality we have generated.')

elif option == 'Text to Action':
    st.header('Demo Video for Text to Action:')
    st.video('sample00_rep00.mp4')
    st.caption('This is a Demo Video displaying the rendering result and motion quality we have generated.')


def Animation(action):
    !sudo apt update
    !sudo apt install ffmpeg
    !wget https://repo.anaconda.com/miniconda/Miniconda3-py37_4.9.2-Linux-x86_64.sh
    !bash Miniconda3-py37_4.9.2-Linux-x86_64.sh -bfp /usr/local
    # Step 2: Set up Conda
    sys.path.append('/usr/local/lib/python3.7/site-packages')
    !conda env create -f environment.yml
    !conda activate mdm
    !python -m spacy download en_core_web_sm
    !pip install git+https://github.com/openai/CLIP.git
    os.chdir('/content/drive/MyDrive/motion-diffusion-model-main')
    print(os.getcwd())
    !bash prepare/download_smpl_files.sh
    !bash prepare/download_glove.sh
    !bash prepare/download_t2m_evaluators.sh
    !pip install smplx
    !pip install matplotlib
    !pip install spacy
    !pip install chumpy
    !pip install typing_extensions==4.7.1
    !pip install torchvision
    !pip install basemap
    !python -m sample.generate --model_path /content/drive/MyDrive/motion-diffusion-model-main/save/humanml_trans_enc_512/model000200000.pt --text_prompt action %s
    !pip install trimesh
    !pip install h5py
    !python -m visualize.render_mesh --input_path /content/drive/MyDrive/motion-diffusion-model-main/save/humanml_trans_enc_512/samples_humanml_trans_enc_512_000200000_seed10_the_person_is_hopping_like_a_rabbit/sample00_rep01.mp4
    st.video('/content/drive/MyDrive/motion-diffusion-model-main/save/humanml_trans_enc_512/samples_humanml_trans_enc_512_000200000_seed10_the_person_is_hopping_like_a_rabbit/sample00_rep01.mp4')



st.markdown('##')
st.header('Create your own animation from text')
action = st.text_input('Enter your action here')

if st.button("Generate Animation"):
    Animation(action)
    st.write('Loading your animation')
else:
    pass

st.write("Find the relevant links about the project here:")
st.write("1. [Colab Notebook for Generating Avatar](https://colab.research.google.com/drive/1KqXGHCWLzFKbKeIIKa6x3AH405nYgD8-)")
st.write("2. [Colab Notebook for Animation/Rendering](https://colab.research.google.com/drive/1GGdgTGtcSLpE6SbE6737kc0SPF3kPvC0)")
st.write("3. [Presentation](https://docs.google.com/presentation/d/1a-apCnnJMW4-ZTVg7bLf8wtuUHkdH21oOcmssEEqULU/edit#slide=id.g2a2bd6182eb_1_0)")
st.write("4. [GitHub](https://github.com/Yzmblog/MonoHuman)")
st.write("5. [Paper by the Authors](https://arxiv.org/abs/2304.02001)")
st.write("6. [Project Page](https://yzmblog.github.io/projects/MonoHuman/)")
```
